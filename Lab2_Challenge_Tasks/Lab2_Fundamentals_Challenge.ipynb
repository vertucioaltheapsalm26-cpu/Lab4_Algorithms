```python
surname = "Vertucio"
SEED_SUM = 7
result = ""

for char in surname:
    if char.lower() in "aeiou":  # Check if it's a vowel
        result += str(SEED_SUM)  # Replace with SEED_SUM
    else:
        result += char           # Keep consonants unchanged

# Print the transformed word
print("Transformed word:", result)
```

```python
ENGINEER = "Vertucio"
loops = SEED_SUM + 3
FirstLetter = ENGINEER[0]
print(FirstLetter)
AsciiVal = ord(FirstLetter)
print(AsciiVal)
verification_hash = (AsciiVal * SEED_SUM) - loops
print(verification_hash)
```

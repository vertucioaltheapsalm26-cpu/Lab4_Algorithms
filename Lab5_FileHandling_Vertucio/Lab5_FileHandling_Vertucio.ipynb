```python
from pathlib import Path
from datetime import datetime
import shutil

# Step 2: Define working directory
# Note: I'm keeping the name consistent with your previous exercises
backup_dir = Path.home() / "Documents" / "Vertucio_Activity_5"
backup_dir.mkdir(exist_ok=True)

# Task 1 & 3: Define a function to write a file with backup and choice of mode
def write_with_backup(filename: str, content: str, mode: str = "w"):
    file_path = backup_dir / filename
    
    # Create backup if file exists
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        # Task 1: Added surname "Vertucio" to the backup filename format
        backup_name = f"{file_path.stem}_Vertucio_backup_{timestamp}{file_path.suffix}"
        backup_path = file_path.with_name(backup_name)
        
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")
    
    # Task 3: Use the 'mode' parameter to decide between overwrite (w) or append (a)
    with open(file_path, mode, encoding="utf-8") as file:
        file.write(content)
        print(f"File saved ({'overwritten' if mode == 'w' else 'appended'}): {file_path.name}")

# Task 2: Function that lists only backups for a given filename
def list_backups_for_file(filename: str):
    print(f"\n--- Backups for {filename} ---")
    stem = Path(filename).stem
    # Using glob to find files that match the specific filename + our backup pattern
    for backup in backup_dir.glob(f"{stem}_Vertucio_backup_*"):
        print("-", backup.name)

# --- Demonstration ---
print("=== File Operations Demo ===")

# Example of overwriting
write_with_backup("demo.txt", "Initial content", mode="w")

# Example of appending (Task 3)
write_with_backup("demo.txt", "\nThis was appended!", mode="a")

# Listing specific backups (Task 2)
list_backups_for_file("demo.txt")
```

```python
from pathlib import Path
from datetime import datetime
import shutil

# Task 1: Change the working directory to Surname_Activity_5
backup_dir = Path.home() / "Documents" / "Vertucio_Activity_5"
backup_dir.mkdir(exist_ok=True)

def file_manager():
    # Ask for filename once at the start
    file_name = input("Enter filename (e.g., notes.txt): ")
    file_path = backup_dir / file_name

    while True:
        print("\n--- MENU ---")
        print("1. Write to file")
        print("2. Append to file")
        print("3. Read file")
        print("4. Backup file")
        print("5. List all backup files") # Task 2: Added 6th option (Logic combined here)
        print("6. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            content = input("Enter content to write:\n")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(content)
            print("File written successfully.")

        elif choice == "2":
            more = input("Enter content to append:\n")
            with open(file_path, "a", encoding="utf-8") as f:
                f.write("\n" + more)
            print("Content appended.")

        elif choice == "3":
            if file_path.exists():
                with open(file_path, "r", encoding="utf-8") as f:
                    print("\nFile Content:\n", f.read())
            else:
                print("File not found.")

        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                # Task 3: Modified backup name to include surname
                backup_file = file_path.with_name(
                    f"{file_path.stem}_Vertucio_backup_{timestamp}{file_path.suffix}"
                )
                shutil.copy2(file_path, backup_file)
                print(f"Backup created: {backup_file.name}")
            else:
                print("Cannot backup. File does not exist.")

        elif choice == "5": # Task 2 Logic: List backups
            print("\n--- All Backups in Directory ---")
            backups = list(backup_dir.glob("*_backup_*"))
            if backups:
                for b in backups:
                    print("-", b.name)
            else:
                print("No backups found.")

        elif choice == "6":
            print("Exiting the file manager.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    file_manager()
```
